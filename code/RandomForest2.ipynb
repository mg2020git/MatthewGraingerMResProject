```python
# Importing libraries
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor # For random forest regression
from sklearn.model_selection import train_test_split # For sp, r2_scorelitting training data and test data
from sklearn.metrics import mean_squared_error, r2_score # For determining the mean squared error and r squared
import matplotlib.pyplot as plt # For plotting importance of RF explanatory variables
import rpy2.robjects as robjects # For running the clean_ASV_table.R script without having to switch kernel
```

```python

clean_asv_script = """

# Importing asv table and meta data into R
asv_table <- read.csv("../data/seqtable_readyforanalysis.csv", sep = "\t")
meta_data <- read.csv("../data/metadata_Time0D-7D-4M_May2022_wJSDpart_ext.csv", sep = "\t")


###################################
#    clean_ASV_table.R  
##################################
# Author: Alberto Pascual-García
# Copyright (c)  Alberto Pascual-García,  2024
# Web:  apascualgarcia.github.io
# Notes added by Matthew Shaun Grainger
# 
# Date: 2024-04-17
# Script Name: name_of_script.R  
# Script Description: This script takes the ASV table
# and metadata, cleans the ASV table removing low
# frequency ASVs and low-depth samples, and then
# gives two options two further select samples:
# --- match.exp = FALSE # if TRUE it will keep only
#                       # samples that were resurrected
# --- exclude.exp = c("4M") # a character vector indicating
#                     # the experiments to exclude, following
#                     # the levels defined in the samples' metadata
#                     # field "Experiment". Defaults to "4M" and it will
#                     # always be removed

clean_ASV_table = function(ASV.table, sample_md,
                           match_exp = FALSE,
                           exclude_exp = c("4M"),
                           nreads = 10000){
  # CLEAN DATA ---------- 
  
  exclude_exp_default = c("4M") # Sets default to exclude 4M samples
  
  # Clean data   ------
  ASV.table=ASV.table[,colSums(ASV.table) > 0] # Gets rid of ASVs with 0 abundance across all samples
  ASV.table=ASV.table[rowSums(ASV.table) > nreads, ] # Gets rid of samples with less than 10,000 reads
  which(is.na(ASV.table))
  
  # .... exclude samples that didn't pass quality control (not present in samples metadata)
  matched=match(row.names(ASV.table),sample_md$sampleid) #attr(data.dist,"Labels"))
  ASV.table=ASV.table[!is.na(matched),] # Gets rid of samples that didn't pass quality control (not in metadata)
  dim(ASV.table)
  
  # ... rebuild the metadata, it may have more samples
  matched=match(row.names(ASV.table),sample_md$sampleid)
  matched=matched[!is.na(matched)]
  sample_md = sample_md[matched, ]
  
  # Exclude data ----
  if((length(exclude_exp) != 0) | (match_exp == TRUE)){
    sample_md_red = sample_md
    
    # --- Always exclude 4M    
    idx = which(sample_md_red$Experiment != exclude_exp_default)
    sample_md_red = sample_md_red[idx, ]
    
    # --- Continue  matching resurrected communities
    if(match_exp == TRUE){ # exclude communities that were not resurrected
      nmatched = 5 # parent community id should appear 5 times
      start_in_exclude = grep("0D", exclude_exp)
      final_in_exclude = grep("7D", exclude_exp)
      nexcluded = length(start_in_exclude)-length(final_in_exclude)
      if(nexcluded == nmatched){
        stop("It doesn't make sense excluding 0D and 7D communities and match_exp = T") 
      }
      nchild = table(sample_md_red$parent)
      idx = which(nchild == nmatched)
      matched = match(sample_md_red$parent, names(nchild)[idx]) 
      sample_md_red = sample_md_red[!is.na(matched), ]
    }
    
    # --- Finally, remove specific experiments or replicates
    if(length(exclude_exp) > 1){ # exclude experiments beyond 4M
      #browser()
      for(level in exclude_exp){
        idx = which(sample_md_red$Experiment != level)
        sample_md_red = sample_md_red[idx, ]
      }
    }
    sample_md = sample_md_red #
    
    # ... look for the positions of remaining samples
    matched=match(sample_md$sampleid,row.names(ASV.table))
    matched=matched[!is.na(matched)]
    
    # ... reshape objects 
    ASV.tmp=ASV.table[matched,] # this tmp file is unnecessary, but easier to debug
    dim(ASV.tmp)
    ASV.table = ASV.tmp # this tmp file is unnecessary, but easier to debug
    ASV.table=ASV.table[,colSums(ASV.table) > 0]
    # ... rebuild the metadata, it may have more samples
    matched=match(row.names(ASV.table),sample_md$sampleid)
    matched=matched[!is.na(matched)]
    sample_md = sample_md[matched, ]
  }
  return(list("ASV.table" = ASV.table,"sample_md" = sample_md))
}

a = clean_ASV_table(ASV.table = asv_table,
                sample_md = meta_data,
                match_exp = TRUE,
                exclude_exp = c("4M", "0D"),
                nreads = 10000)
asv_table = a[[1]]
meta_data = a[[2]]
write.csv(asv_table, "../data/cleaned_asv_table.csv")
write.csv(meta_data, "../data/cleaned_meta_data.csv")


"""

robjects.r(clean_asv_script) # Running the clean ASV R scipt within Python
```

```python
# Importing data
asv_table = pd.read_csv('../data/cleaned_asv_table.csv', index_col=0)
cluster_data = pd.read_csv("../data/max_tot_ext_network_table.tsv", delimiter='\t')
meta_data = pd.read_csv("../data/cleaned_meta_data.csv")
function_data = pd.read_csv("../data/20151016_Functions_remainder.csv")
```

## 2 - Modifying function data
Adding .# to specify replicates and changing name to sampleid. Converting units and adding addtional normalised columns for relevant variables:
- Converting mgCO2.7 from miligrams to micrograms and ATP7 from nanomolar to micromolar.
- Normalising ATP (ATP7) and xylosidase (mX7) by the number of cells (CPM7).
- Taking the logarithm of these normalised ATP7 and mX7, as well as the non-normalised versions.

```python
# Converting units
function_data['mgCO2.7'] = function_data['mgCO2.7'] * 1000 # Converting unit
function_data.rename(columns={'mgCO2.7': 'µgCO2.7'}, inplace=True) # Chaning name
function_data['ATP7'] = function_data['ATP7'] / 1000 # Converting unit

# Normalising
function_data['ATP7.norm'] = function_data['ATP7'] / function_data['CPM7'] # Normalised ATP
function_data['mX7.norm'] = function_data['mX7'] / function_data['CPM7'] # Normalised cell count

# Taking the logarithms
function_data['log.ATP7.norm'] = np.log10(function_data['ATP7.norm'] + 0.00001)
function_data['log.ATP7'] = np.log10(function_data['ATP7'] + 0.00001)
function_data['log.mX7.norm'] = np.log10(function_data['mX7.norm'] + 0.00001)
function_data['log.mX7'] = np.log10(function_data['mX7'] + 0.00001)
```

```python
def specify_replicate(row): # Function to specify replicate of final sample
    if row['Replicate'] == 1: # If the replicate column has a value of 1
        return f"{row['Community']}.1" # Adding .1 to the community column sampleid
    elif row['Replicate'] == 2:
        return f"{row['Community']}.2"
    elif row['Replicate'] == 3:
        return f"{row['Community']}.3"
    elif row['Replicate'] == 4:
        return f"{row['Community']}.4"

function_data['Community'] = function_data.apply(specify_replicate, axis=1) # Applying the function to the function_data
function_data = function_data.rename(columns={'Community': 'sampleid'}) # Changing the name of the communtiy column to sampleid
```

## 2 - Rarefying ASV table to 10,000 reads
Rarefying samples with over 10,000 reads down to 10,000 reads. (NOTE: The rows of the asv_table and the meta_data are aligned.)

```python
#### Filtering out samples with less than 10,000 reads, removing experiment 4M samples (different experiment), and removing starting samples

#asv_table['reads'] = asv_table.sum(axis=1) # Column with number of reads for each sample
#asv_table.reset_index(inplace=True) # Making the sample ID into a column for the ASV table
#asv_table.rename(columns={'index': 'sampleid'}, inplace=True) # renaming this new column to 'sampleid'


#indices = asv_table.index[asv_table['reads'] < 10000].tolist() # Indices of samples with less than 10,000 reads
#asv_table = asv_table.drop(indices)
#meta_data = meta_data.drop(indices)

#indices = meta_data.index[meta_data['Experiment'] == '4M'].tolist() # Experiment 4M samples
#asv_table = asv_table.drop(indices)
#meta_data = meta_data.drop(indices)

#indices = meta_data.index[meta_data['replicate'] == 'Rep0'].tolist() # Starting samples (before cryopreservation)
#asv_table = asv_table.drop(indices)
#meta_data = meta_data.drop(indices)

#asv_table.set_index('sampleid', inplace=True) # Making sampleid back into row names instead of columns


#### Setting rarefaction depth to 10,000
rarefaction_depth = 10000

# Function to rarefy a sample
def rarefy_vector(v, depth):
    non_zero_indices = np.nonzero(v)[0] # Identify indices of non-zero ASVs, as these are the ones with reads
    probabilities = v[non_zero_indices] / v.sum() # Calculating probabilities of selecting each non-zero ASV
    subsampled = np.random.choice(non_zero_indices, size=depth, replace=True, p=probabilities)  # Randomly subsample the non-zero indices to depth
    rarefied = np.zeros_like(v) # New vector with the same size as the original
    np.add.at(rarefied, subsampled, 1) # Increment the counts in the rarefied vector based on the subsampling
    return rarefied

# Function to rarefy the ASV table (where rows are samples)
def rarefy_table(df, depth):
    rarefied_data = df.apply(lambda x: rarefy_vector(x.values, depth), axis=1) # Applies function that rarefies a sample to each row of the asv table
    rarefied_df = pd.DataFrame(rarefied_data.tolist(), index=df.index, columns=df.columns) # 
    return rarefied_df

# Rarefying the asv table
asv_table = rarefy_table(asv_table, rarefaction_depth)

# Getting rid of reads column
#asv_table.drop(columns=['reads'], inplace=True)
```

## 4 - Creating data frame with abundances of each cluster in each sample

```python
# Transposing so that there is a column for ASVs
cluster_table = asv_table.transpose()

# Merging with the cluster data by ASV, such that there is a single cluster column
cluster_table = cluster_table.merge(cluster_data, left_index=True, right_on='ASV') # Merging by ASV
cluster_table.drop(['ASV', 'Set.x'], axis=1, inplace=True) # Getting rid of unnecessary columns
# Now have df with columns for samples and a column for cluster

# Grouping by cluster and transposing so that each cluster is a column with an abundance in each sample
cluster_table = cluster_table.groupby('functionInk').sum() # Summing all of the rows that have the same cluster value
cluster_table  = cluster_table.transpose() # Transposing so that clusters are columns
```

## 5 - Creating metavariable data frames

```python

# Getting rid of unnecessary metavariables
meta_data= meta_data[['sampleid','parent', 'Location', 'partition']] # Only kept parent (to group replicates by tree hole) and location (to group tree holes by location).

# Bringing sampleid column into asv_table for merging
asv_table.reset_index(inplace=True) # Making the sample ID into a column for the ASV table
asv_table.rename(columns={'index': 'sampleid'}, inplace=True) # renaming this new column to 'sampleid'

# Bringing sampleid column into cluster_table for merging
cluster_table.reset_index(inplace=True)
cluster_table.rename(columns={'index': 'sampleid'}, inplace=True)

# Merging with meta data
asv_table_metavar = pd.merge(asv_table, meta_data, on='sampleid', how='inner')
cluster_table_metavar = pd.merge(cluster_table, meta_data, on='sampleid', how='inner')

# Encoding non-numeric metavariables (can only use numeric for RF)
asv_table_metavar = pd.get_dummies(asv_table_metavar, columns=['parent', 'Location', 'partition'])
cluster_table_metavar = pd.get_dummies(cluster_table_metavar, columns=['parent', 'Location', 'partition'])
```

## 6 - Aligning function data with the 4 explanatory data frames

All 4 explanatory data frames have the same number of rows, and these should be in the same order.

```python

# Merging tables with function data
merged_table_1 = pd.merge(asv_table, function_data, on='sampleid', how='inner')
merged_table_2 = pd.merge(asv_table_metavar, function_data, on='sampleid', how='inner')
merged_table_3 = pd.merge(cluster_table, function_data, on='sampleid', how='inner')
merged_table_4 = pd.merge(cluster_table_metavar, function_data, on='sampleid', how='inner')

# Unmerging
columns_to_drop = list(function_data.columns.values) # List of columns from function_data
function_data = merged_table_1[columns_to_drop] # Only keeping function data columns
columns_to_drop.remove('sampleid') # Removing sampleid so that it isn't dropped
asv_table = merged_table_1.drop((columns_to_drop), axis = 1) # Dropping these columns to get asv table
asv_table_metavar = merged_table_2.drop((columns_to_drop), axis = 1)
cluster_table = merged_table_3.drop((columns_to_drop), axis = 1)
cluster_table_metavar = merged_table_4.drop((columns_to_drop), axis = 1)

# Sorting data frames by sampleid to ensure same order of rows
asv_table = asv_table.sort_values(by='sampleid')
asv_table_metavar = asv_table_metavar.sort_values(by='sampleid')
cluster_table = cluster_table.sort_values(by='sampleid')
cluster_table_metavar = cluster_table_metavar.sort_values(by='sampleid')
function_data = function_data.sort_values(by='sampleid')

# Dropping sampleid from explanatory data frames
asv_table = asv_table.drop(('sampleid'), axis = 1)
asv_table_metavar = asv_table_metavar.drop(('sampleid'), axis = 1)
cluster_table_metavar = cluster_table_metavar.drop(('sampleid'), axis = 1)
cluster_table = cluster_table.drop(('sampleid'), axis = 1)

# Converting column names to strings
cluster_table_metavar.columns = cluster_table_metavar.columns.astype(str)
cluster_table.columns = cluster_table.columns.astype(str)
```

## 7 - Creating response data frames
The 4 explanatory data frames created above seem to be aligned (each sample seems to be represented by the same row across all 4 dataframes).
Now response data frames for:
- log normal xylo 
- log normal ATP
- 

```python

# Response data frames
log_normalised_xylo = function_data['log.mX7.norm'] # The log of xylosidase normalised by cell count
log_normalised_atp = function_data['log.ATP7.norm'] # The log of ATP normalised by cell count
```

## 8 - Random Forest Regressions

Need to perform 2 RFs for each response: one with ASVs as explanatory and one with clusters as explanatory.
RFs:
- log_normalised_xylo * ASV
- log_normalised_xylo * cluster
- log_normalised_atp * ASV
- log_normalised_atp * cluster

```python
####### ASV : Xylo

# Split the data between training data and testing data
X_train, X_test, y_train, y_test = train_test_split(asv_table_metavar, log_normalised_xylo, test_size=0.2, random_state=42)

# Initialize and train the Random Forest Regressor
asv_xylo_rf = RandomForestRegressor(n_estimators=100, random_state=42) # Initialise. n_estimators is no. trees, random_state is seed for reproducibility
asv_xylo_rf.fit(X_train, y_train) # Trains the model

# Making predictions
y_pred = asv_xylo_rf.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}') # Less is better, higher is worse. DIfference between predicted and observed.
print(f'R-squared: {r2}') # Higher means better fit. Prop of variance in dependent explained by independent.

# Extract feature importances
feature_importances = asv_xylo_rf.feature_importances_

# Create a DataFrame for better visualization
importance_df = pd.DataFrame({
    'Feature': X_train.columns,
    'Importance': feature_importances
})
importance_df = importance_df.sort_values(by='Importance', ascending=False) # Sort values by importance
importance_df = importance_df[importance_df['Importance'] > 0] # Get rid of features with 0 importance
asv_xylo_importance_df = importance_df
asv_xylo_importance_df = asv_xylo_importance_df[asv_xylo_importance_df['Importance'] > 0.01] # Get rid of features with very low importance

# Plot the feature importances
plt.figure(figsize=(10, 8))
plt.barh(asv_xylo_importance_df['Feature'], asv_xylo_importance_df['Importance'])
plt.xlabel('Importance')
plt.ylabel('Feature')
plt.title('Feature Importances in ASV-Xylosidase Random Forest Regressor')
plt.gca().invert_yaxis()

# Save the plot as a PNG file
#plt.savefig('../results/feature_importances_asv_xylo.png', bbox_inches='tight')

plt.show()

```
